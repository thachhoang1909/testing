## Q. 3: 
We have an array of n elements A[1..n]. This array contains n different numbers from 0 to
n. Given that there are totally n + 1 numbers from 0 to n, there is a missing number. Write a
function to find this missing number in O(n) time.

```
def find_missing(A):
  
  # create a list with n+1 zero values
  list_num = [0]*(n+1)
  
  ## loop over list A and flag the position of each element from A in list_num to 1
  for i in range(n+1): ## range function return the list from 0 to n
    num = A[i]
    list_num[num] = 1
  
  ## loop over list_num to find the position with 0 value
  for i in range(n+2): ## range function return the list from 0 to n+1
    if list_num[i] == 0:
      print("Missing number: ", i)
```

## Q. 6: 
We have three identical six-sided dices. We role one dice first and the remaining two dices
after that. What is the probability that the point obtained in the first roll is greater than sum of
the points obtained in the second roll. \\

- Solution

The first dice must be 3, 4, 5 or 6 to greater than the sum of the second roll which minimum is 2 (2nd dice = 1, 3rd dice = 1)

Let's list some possible outcomes to see if there is a logic behind:\
Case 1st dice $d_1$= 3: (1, 1)\
Case 1st dice $d_1$= 4: (1, 1) - (1,  2)\
Case 1st dice $d_1$= 5: (1, 1) - (1, 2) - (1, 3) - (2, 1) - (2, 2) - (3, 1)\
Case 1st dice $d_1$= 6: (1, 1) - (1, 2) - (1, 3) - (1, 4) - (2, 1) - (2, 2) - (2, 3) - (3, 1) - (3, 2)

Let $d_1, d_2$ is the outcome of the first and second dice, respectively.

As can be seen, the number of possible outcomes of 2nd dice equal to $d_1 - 2$ and the number of possible outcomes of 3rd dice equal to $d_1 - (d_2  + 1)$  

The probability of each outcome of a dice is $\frac{1}{6}$
Then we have the probability that the point obtained in the first roll is greater than sum of
the points obtained in the second roll: 
$$p = \frac{1}{6} \sum_{d_1=3}^{6} \ \frac{1}{6} \sum_{d_2=1}^{d_1-2} \frac{1}{6}(d_1 - (d_2 + 1))$$

$$p = \frac{1}{6^3} \sum_{d_1=3}^{6} \ \sum_{d_2=1}^{d_1-2} (d_1 - (d_2 + 1))$$

$$p = 0.09259259259259259$$

```
## Code using to calculate p
temp = 0
for i in range(3, 6+1):
  for j in range(1, i-2+1):
    temp += (i- j - 1)

result = temp/(6**3)
result
```

## Q. 7: 
Given a Bayesian Network as in Figure 1

A -> B -> C, B -> C

The above Bayesian Network has the following probabilities

• P(A) = 0.6\
• P(B|A) = 0.7, P(B|\bar{A}) = 0.3\
• P(C|B) = 0.2, P(C|\bar{B}) = 0.8\
• P(D|B) = 0.1, P(D|\bar{B}) = 0.4

Calculate the probability of $P(\bar{C})$?


$P(\bar{C}) = 1 - P(C)$

$P(C) =\sum_A \sum_B P(A) *P(B | A) * P(C|B) $


$ \ \ \ \ \ \ \ \ \ = \sum_A P(A) \sum_B  P(B | A) * P(C|B)  $

$ \ \ \ \ \ \ \ \ \ =  \sum_A P(A) \  [P(B | A) * P(C|B) + P(\bar{B} | A) * P(C|\bar{B})]  $

$ \ \ \ \ \ \ \ \ \ =   P(A) \  [P(B | A) * P(C|B) + P(\bar{B} | A) * P(C|\bar{B})] + P(\bar{A}) \  [P(B | \bar{A}) * P(C|B) + P(\bar{B} | \bar{A}) * P(C|\bar{B})]$

$  \ \ \ \ \ \ \ \ \ = 0.6*(0.7*0.2 + 0.3*0.8) + 0.4*(0.3*0.2 + 0.7*0.8)$\
$  \ \ \ \ \ \ \ \ \ = 0.476$

$\rightarrow P(\bar{C}) = 1 - P(C) = 1 - 0.476 = 0.524$ 

Where:

$P(\bar{B} | A) = 1 - P(B | A) = 0.3$\
$P(\bar{A}) = 1 - P(A) = 0.4$\
$P(\bar{B} | \bar{A})  = 1 - P(B | \bar{A})  = 0.7$

We can use the t-test to see if there is a signification between the mean ridership of rainy day versus non-rainy day. But there are two assumptions must be true: 
- **independent observations**: the data is the record of ridership on rainy days and non-rainy days in the real world so it is independent (True)
- **normally distribution** or **large samples** or **small sample - 2 sides test**: the ENTRIESn_hourly is not normal distribution and the sample data is not large. So, I will use 2 tailed t-test.



*   Null hypothesis: mean of ENTRIESn_hourly (new_rain=1) == mean of ENTRIESn_hourly (new_rain=0)
*   Alternative hypothesis: mean of ENTRIESn_hourly (new_rain=1) != mean of ENTRIESn_hourly (new_rain=0) 
* Significance Level: 0.05


```
rain_day = nyc_subway[['new_rain', 'ENTRIESn_hourly']][nyc_subway.new_rain==1]
nonrain_day = nyc_subway[['new_rain', 'ENTRIESn_hourly']][nyc_subway.new_rain==0]

ttest_ind(rain_day, nonrain_day)
```

$$ p_{value} = 0.00763278 < \alpha $$
-> Reject null hypothesis, so can conclude that rainy day affects the ridership.
